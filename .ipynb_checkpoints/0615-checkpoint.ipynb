```python
import numpy as np
import matplotlib.pylab as plt
rng = np.random.RandomState(1)

x = 30*rng.rand(100) +150

y = 1.5 * x - 180

y += rng.randn(100)

plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis],y)
print(model.coef_, model.intercept_)
xfit = np.linspace(150, 180, 1000)
yfit= model.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit, color ="red")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

rng = np.random.RandomState(1)

x = 200 * rng.rand(100)+ 200
y = -3 * x + 200
y += (rng.randn(100)*30)

plt.scatter(x,y)

model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis],y)

xfit = np.linspace(200, 300, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.plot(xfit, yfit, color="red")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
%matplotlib inline
```

```python
insurance_data = pd.read_csv("/home/kiro04/Documents/robot_ws/i_robot_ws/data/insurance.csv")
print(insurance_data)
```

```python
sns.pairplot(insurance_data)
```

```python
non_numeric_columns = insurance_data.select_dtypes(exclude=np.number).columns
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for i in non_numeric_columns:
    insurance_data[i] = label_encoder.fit_transform(insurance_data[i])
print(insurance_data)
```

```python

insurance_data.corr()
```

```python
plt.figure()
sns.heatmap(insurance_data.corr(), annot = True, cmap="PiYG") #  ['PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu', 'RdYlBu','RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic'])
plt.show()
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
print(insurance_data)

num_vars = ["age", "bmi","charges"]
insurance_data[num_vars] = min_max_scaler.fit_transform(insurance_data[num_vars])
print(insurance_data)
```

```python
x_input_data = insurance_data.to_numpy(dtype = np.float32)[:1000,:-1]
y_answer_data = insurance_data.to_numpy(dtype = np.float32)[:1000,-1]
print(x_input_data)
print(y_answer_data)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_input_data, y_answer_data)
```

```python
x_test= insurance_data.to_numpy(dtype=np.float32)[:1000,:-1]
y_test= insurance_data.to_numpy(dtype=np.float32)[:1000,-1,np.newaxis]
y_pedict = model.predict(x_test)
fig = plt.figure()
plt.scatter(y_test, y_pedict,alpha=.5,c=["r"])
plt.show()
```
